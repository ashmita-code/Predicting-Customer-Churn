```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = df.copy()
df.drop(columns=["tenure_group", "tenure_year_group", "tenure_years"], errors="ignore", inplace=True)




df["tenure_years"] = (df["tenure"] / 12).round(1)


bins = np.arange(0, 7.5, 1)
labels = [f"{i}-{i+1}" for i in range(7)]
df["tenure_year_group"] = pd.cut(
    df["tenure_years"], bins=bins, labels=labels, include_lowest=True
)


churn_rate_years = (
    df.groupby("tenure_year_group")["Churn"]
      .apply(lambda x: (x == "Yes").mean() * 100)
      .reset_index(name="Churn Rate (%)")
)


plt.figure(figsize=(8,5))
sns.lineplot(
    data=churn_rate_years,
    x="tenure_year_group",
    y="Churn Rate (%)",
    marker="o"
)
plt.title("Churn Rate by Tenure (Years)")
plt.xlabel("Tenure Range (Years)")
plt.ylabel("Churn Rate (%)")
plt.grid(True)
plt.show()

churn_rate_years
```

🧠 Insight: Churn Rate by Tenure (Years)

Customers in their first year (0–1 years) have the highest churn rate (~45–50%).

After 2 years, churn drops to roughly 25–30%.

Beyond 3 years, churn remains below 20%, and by 6 years, it’s in the single digits.

Eda done, now modeling

```python


data = df.copy()


data = data.drop(columns=["tenure_group", "tenure_year_group", "tenure_years"], errors="ignore")


binary_cols = ['gender','Partner','Dependents','PhoneService','PaperlessBilling','Churn']
for col in binary_cols:
    data[col] = data[col].map({'Yes':1, 'No':0, 'Male':1, 'Female':0})


multi_cols = ['MultipleLines','InternetService','OnlineSecurity','OnlineBackup',
              'DeviceProtection','TechSupport','StreamingTV','StreamingMovies',
              'Contract','PaymentMethod']
data = pd.get_dummies(data, columns=multi_cols, drop_first=True)


data = data.drop(columns=['customerID'])

print("Prepared data shape:", data.shape)
data.head(3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd


if "SeniorCitizen" in data.columns and data["SeniorCitizen"].dtype == "object":
    data["SeniorCitizen"] = data["SeniorCitizen"].map({"Yes": 1, "No": 0}).astype(int)


bool_cols = data.select_dtypes(include=["bool"]).columns
if len(bool_cols) > 0:
    data[bool_cols] = data[bool_cols].astype(int)


non_num = data.drop(columns=["Churn"]).select_dtypes(exclude=[np.number]).columns.tolist()
print("Non-numeric feature columns (should be empty):", non_num)


X = data.drop("Churn", axis=1)
y = data["Churn"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_test  = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score


log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)


y_pred = log_reg.predict(X_test)


print("🔹 Logistic Regression Results 🔹")
print("Accuracy:", round(accuracy_score(y_test, y_pred), 3))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier


rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)


y_pred_rf = rf.predict(X_test)


print("🌲 Random Forest Results 🌲")
print("Accuracy:", round(accuracy_score(y_test, y_pred_rf), 3))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


feat_importances = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': rf.feature_importances_
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(data=feat_importances.head(10), x='Importance', y='Feature', palette='coolwarm')
plt.title("Top 10 Features Influencing Churn (Random Forest)")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()
```

nterpretation Summary

Retention drivers: long tenure, contract type, tech support, online security.

Risk factors: short tenure, high monthly charge, electronic payment, fiber optic service.

SMOTE + XGBoost

```python


from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

print("Before SMOTE class counts:", y_train.value_counts().to_dict())
print("After  SMOTE class counts:", y_train_sm.value_counts().to_dict())
```

```python


from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)

xgb.fit(X_train_sm, y_train_sm)

y_prob = xgb.predict_proba(X_test)[:,1]
y_pred_adj = (y_prob > 0.4).astype(int)

print("🚀 XGBoost (Threshold = 0.4) Results 🚀")
print("Accuracy:", round(accuracy_score(y_test, y_pred_adj), 3))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_adj))
print("\nClassification Report:\n", classification_report(y_test, y_pred_adj))
```

⚖️ Step-by-step logic
1. What accuracy really measures

Accuracy = (Correct Predictions) / (All Predictions)

In imbalanced datasets (like churn), most customers are non-churners.
So even if your model predicts “No churn” for almost everyone, you’ll still get high accuracy, but it’s misleading because you miss many true churners.

2. Why recall matters more here

In churn prediction, the business goal is to catch as many churners as possible, so the company can intervene (discounts, retention calls, etc.).

That’s what recall measures:

Recall = (True Churners correctly predicted) / (All Actual Churners)

Before (threshold = 0.5) → Recall = 0.65 → Model caught 65 % of churners

After (threshold = 0.4) → Recall = 0.72 ✅ → Model caught 72 % of churners

That’s 7 % more churners saved — real money for the company.

```python
from sklearn.metrics import precision_recall_curve, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
y_prob = xgb.predict_proba(X_test)[:,1]
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)


f1_scores = 2 * (precision * recall) / (precision + recall)
best_threshold = thresholds[np.argmax(f1_scores)]

print("Best Threshold (max F1):", round(best_threshold, 3))
```

```python
plt.figure(figsize=(7,5))
plt.plot(thresholds, precision[:-1], label='Precision', color='green')
plt.plot(thresholds, recall[:-1], label='Recall', color='red')
plt.plot(thresholds, f1_scores[:-1], label='F1-score', color='blue')
plt.axvline(x=best_threshold, color='black', linestyle='--', label=f'Best = {best_threshold:.2f}')
plt.title("Precision–Recall vs Threshold")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.legend()
plt.show()
```

revaluting the modl

```python


from sklearn.metrics import classification_report, confusion_matrix, accuracy_score


y_prob = xgb.predict_proba(X_test)[:, 1]


optimal_threshold = 0.447
y_pred_opt = (y_prob > optimal_threshold).astype(int)


print("🚀 XGBoost Results (Optimized Threshold = 0.447) 🚀")
print("Accuracy:", round(accuracy_score(y_test, y_pred_opt), 3))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_opt))
print("\nClassification Report:\n", classification_report(y_test, y_pred_opt))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def metrics_for(th):
    y_pred = (y_prob > th).astype(int)
    return [
        round(accuracy_score(y_test, y_pred), 3),
        round(precision_score(y_test, y_pred), 3),
        round(recall_score(y_test, y_pred), 3),
        round(f1_score(y_test, y_pred), 3),
    ]

labels = ["Accuracy", "Precision (Churn=1)", "Recall (Churn=1)", "F1-score (Churn=1)"]
vals05   = metrics_for(0.5)
vals0447 = metrics_for(0.447)

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(8,5))
b1 = ax.bar(x - width/2, vals05,   width, label='Threshold 0.5',   color='#4a90e2')
b2 = ax.bar(x + width/2, vals0447, width, label='Threshold 0.447', color='#e67e22')

for bars in (b1, b2):
    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x()+bar.get_width()/2, h+0.01, f"{h:.2f}", ha='center', va='bottom', fontsize=9)

ax.set_ylabel('Score')
ax.set_title('Threshold Comparison: Default (0.5) vs Optimized (0.447)')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=10)
ax.set_ylim(0.4, 1.0)
ax.legend()
plt.tight_layout()
plt.show()
```

“Tuning the decision threshold from 0.50 to 0.447 improves recall by 4 pts (65→69%) with negligible impact on F1, enabling earlier detection of at-risk customers.”
